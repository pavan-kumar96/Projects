### Missing Data Visualization

```python
mg.matrix(data)
plt.title("Missing data Distribution")
```

From the above plot the white spaces represent the no. of missing values in the data

### Missinng data Information

```python
plt.figure(figsize=(20,10))
data.isna().sum().plot(kind = 'bar', fontsize = 15)
plt.ylabel('count of missing values')
plt.xlabel('Column names')
plt.title("Missing Data Count")
```

### Gender Distribution

```python

plt.figure(figsize=(15,10))
explode = [0,0.1]
data.sex.value_counts().plot(kind = 'pie', autopct ='%.2f', explode =  explode, fontsize = 20)
plt.legend()
plt.title("Gender Distribution")
```

From the above plot we can say that Hepatatis is found more in "FEMALES" when compare to "MALES"

### Prediction approach

from the above table [alk_phosphate, sgot,protime] are having the high standard deviation meaning the data is not normally distributed

all the above mentioned features are slightly right skewed

```python
col = ['age', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime']
for c in col:
    new = data[col].skew()
new.plot(kind = "bar")
plt.title("Skewess distribution in the model")
```
